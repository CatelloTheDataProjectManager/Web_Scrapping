## Scrapping_Tableau_Wikipédia

https://fr.wikipedia.org/wiki/Liste_des_pays_par_PIB_nominal

```python
import pandas as pd
from bs4 import BeautifulSoup
import requests
```

```python
url = "https://fr.wikipedia.org/wiki/Liste_des_pays_par_PIB_nominal"

page = requests.get(url)
```

```python
soup = BeautifulSoup(page.content, "html")
```

```python
table = soup.find("table", {"class": "wikitable"})

# Extraction des en-têtes de colonnes
headers = [th.text for th in table.find_all("th")]

# Extraction des données de chaque ligne
data = []
for row in table.find_all("tr")[1:]:
    rowData = [td.text for td in row.find_all("td")]
    data.append(rowData)

# Création du dataframe
df = pd.DataFrame(data, columns=headers)
```

### Scrapping_Tableau_NBA

```python
url = "https://www.basketball-reference.com/leagues/NBA_2022_per_game.html"

page = requests.get(url)
```

```python
soup = BeautifulSoup(page.content, "html")
```

```python
table = soup.find("table", {"id": "per_game_stats"})

# Extraction des en-têtes de colonnes
headers = [th.text for th in table.find_all("th")[:29]]

# Extraction des données de chaque ligne
data = []
for row in table.find_all("tr")[1:]:
    rowData = [td.text for td in row.find_all("td")]
    data.append(rowData)

# Création du dataframe
df = pd.DataFrame(data, columns=headers)
```
