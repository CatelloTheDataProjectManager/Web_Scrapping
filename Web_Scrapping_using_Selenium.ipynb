## Selenium

```python
from selenium import webdriver
from selenium.webdriver.common.by import By

# Set the path to your ChromeDriver executable
driver_path = "C:/Users/Catello/Desktop/chromedriver-win32/chromedriver.exe"

# Create a new Service object
service = webdriver.chrome.service.Service(executable_path=driver_path)

# Create a new instance of the Chrome browser
driver = webdriver.Chrome(service=service)
```

### Publications scientifiques DBLP

```python
# Navigate to a webpage
driver.get("https://dblp.org/pid/07/1501.html")
```

#### titles

```python
# Find all elements with class name "title"
title_elements = driver.find_elements(By.CLASS_NAME, "title")

# Extract the text of each element and store it in a list
titles = [element.text for element in title_elements]
```

```python
len(titles)
```

#### authors

```python
# Find all elements with class name "author"
author_elements = driver.find_elements(By.XPATH, '//li[@class="entry article toc"]//span[@itemprop="name"][ancestor::span[@itemprop="author"]]')

# Extract the text of each element and store it in a list
authors = [element.text for element in author_elements]
```

```python
len(authors)
```

```python
authors
```

```python
# Find all elements with class name "entry article toc"
article_elements = driver.find_elements(By.CSS_SELECTOR, 'li.entry.article.toc')

# Create an empty list to store the titles and authors of each article
articles = []

# Iterate over each article element
for article in article_elements:
    # Find the title and authors of the current article
    title = article.find_element(By.CSS_SELECTOR, 'cite.data span.title').text
    authors = [author.text for author in article.find_elements(By.CSS_SELECTOR, 'cite.data span[itemprop="name"]')]

    # Store the title and authors in a dictionary and append it to the list of articles
    articles.append({'title': title, 'authors': authors})

# Print the list of articles (for verification purposes)
print(articles)
```

#### Final Code

```python
from pandas import DataFrame

# Find all elements with class name "entry article toc"
article_elements = driver.find_elements(By.CSS_SELECTOR, 'li.entry.article, li.entry.inproceedings, li.entry.incollection, li.entry.proceedings, li.entry.informal, li.entry.toc')

# Create an empty list to store the titles, authors, and dates of each article
articles = []

# Iterate over each article element
for article_element in article_elements:
    # Find the title, authors, and date of the current article
    title = article_element.find_element(By.CSS_SELECTOR, 'cite.data span.title').text
    authors = [author.text for author in article_element.find_elements(By.CSS_SELECTOR, 'cite.data span[itemprop="name"]')]
    date = article_element.find_element(By.CSS_SELECTOR, 'cite.data span[itemprop="datePublished"]').text

    # Store the title, authors, and date in a dictionary and append it to the list of articles
    articles.append({'title': title, 'authors': authors, 'date': date})

# Create a DataFrame from the list of articles
df = DataFrame(articles)

# Print the DataFrame (for verification purposes)
print(df)
```
